importing necessary libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

reading the image

```python
image = cv2.imread("apple.jpg")
cv2_imshow(image)
```

sharpening function

```python
def image_sharpening(image):
  sharp_kernel = np.array([[0,-1,0],[-1,5,-1] , [0,-1,0]])
  sharpened_image = cv2.filter2D(image , -1, sharp_kernel)
  cv2_imshow(sharpened_image)
```

showing the sharp image

```python
image_sharpening(image)
```

blurring function

```python
def image_blurring(image):
  blurr_kernel = np.ones((15,15),np.float32)/225
  blurred_image = image
  for _ in range(15):
    blurred_image = cv2.filter2D(blurred_image , -1 , blurr_kernel)
  cv2_imshow(blurred_image)
```

showing the blurred image

```python
image_blurring(image)
```

enhancing the contrast

```python
def enhance_contrast(image):
  gray_image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
  enhanced_image = cv2.equalizeHist(gray_image)
  return enhanced_image
```

function to higlight the edges


```python
def edge_highlighting(image):
  gray_image= enhance_contrast(image)
  #gray_image = cv2.GaussianBlur(gray_image , (5,5) , 0)
  kernel = np.array([[0,1,0] , [1,-4,1] , [0,1,0]])
  output = cv2.filter2D(gray_image , -1 , kernel)
  output = np.absolute(output)
  output = cv2.normalize(output , None , 0 ,255 , cv2.NORM_MINMAX)
  output = np.uint8(output)
  _,  output = cv2.threshold(output, 50 , 255 , cv2.THRESH_BINARY)
  cv2_imshow(output)
```

showing the result after edges highlighting

```python
edge_highlighting(image)
```
