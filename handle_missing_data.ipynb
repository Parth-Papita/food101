```python
import pandas as pd

```

```python
df = pd.read_csv("IITK_Courses_Report.csv")
df.head(25)
```

```python
temp = df.select_dtypes(exclude = ['int' , 'float'])
temp
```

```python
for items in temp.columns:
  temp2 = df[df[items].isnull()]
  for val in temp2.index:
    df.drop(val , inplace = True)
df.reset_index(inplace = True , drop = True)
df.head(25)
```

```python
temp = df.select_dtypes(include=['int' , 'float'])
temp
```

```python
for items in temp:
  temp2 = df[df[items].isnull()]
  for val in temp2.index:
     df.loc[val , items] = round((df.loc[val-1,items] * df.loc[val+1,items])**0.5 , 2)
df.head(25)
```

```python
df.to_csv('cleaned_IITK_Courses_Report.csv' , index = False)
```
